### Find total buisness activity 

```python
yelp_api = YelpAPI(key, c_secret, token, t_secret)

def query_yelp_bound_count(bnds):
    cnt = yelp_api.search_query(bounds=bnds, offset=0, limit=1)['total']
    return [int(cnt), bnds]

def generate_query_grid(lat1, lon1, 
                        lat2, lon2,
                        gridspace = 15.0):
    gridspacelat = abs(lat1-lat2)/gridspace
    gridspacelon = abs(lon2-lon1)/gridspace
    gridstrs = []
    gridpts = []    
    for la in np.arange(lat1, lat2, gridspacelat):
        for lo in np.arange(lon2, lon1, gridspacelon):
            bnds = (str(la)+', '+str(lo)+'|'
                    +str(la+gridspacelat)+', '+str(lo-gridspacelon))
            gridpts.append([(la+la+gridspacelat)*0.5,(lo+lo-gridspacelon)*0.5])
            gridstrs.append(bnds)
    print 'total generated by grid = {}'.format(gridspace*gridspace)
    return gridstrs, gridpts

def parallel_query(params):
    pool = multiprocessing.Pool(4)
    result = pool.map(query_yelp_bound_count, params)
    pool.close()
    pool.join()
    return result

if __name__ == "__main__":
    bnds, grid = generate_query_grid(gridspace = 65)
    grid = np.array(grid, dtype='float')
    grid_population = parallel_query(bnds)
    print 'done'
    grid_population = np.array(grid_population, dtype='object')
    grid_population[:,0] = grid_population[:,0].astype(int)
#    temp = pd.DataFrame(grid_population)
#    temp.to_csv('yelp_sig_grid_population.csv',index=False)
```
